```python
from transformers import pipeline
classifier=pipeline(task="sentiment-analysis",model="facebook/bart-large-mnli")
result=classifier("i was not happy with the last mission")
print(result)
```

```python
from transformers import pipeline
classifier=pipeline(task="text-generation",model="distilbert/distilgpt2")
generated_text=classifier("Today is a rainy day in Kerala")
print(generated_text[0]['generated_text'])
```

```python
from transformers import pipeline
qn_model=pipeline(task="question-answering",model="distilbert/distilbert-base-cased-distilled-squad")
question="what is tha capital of India"
result=qn_model(question=question,context="share with my students")
print(result)
```

```python
import arxiv
import pandas as pd
query="Artificial Intelligence"
```

```python
search=arxiv.Search(query=query,max_results=10,sort_by=arxiv.SortCriterion.SubmittedDate)
```

```python
papers=[]
for i in search.results():
 papers.append({
     'Title':i.title,
     'Authors':i.authors,
     'Summary':i.summary,
     'Link':i.links,
     'Published':i.published})
```

```python
df=pd.DataFrame(papers)
pd.set_option('display.max_colwidth',None)
df.head(10)
```

```python
df
```

```python
abstract=df['Summary'][0]
```

```python
from transformers import pipeline
summarizer=pipeline(task="summarization",model="facebook/bart-large-cnn")
```

```python
result=summarizer(abstract)
```
